```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),index=ts.index,
                  columns = ['A','B','C','D'])
df = df.cumsum() #shangmian suijishu qiuhe
```

```python
plt.figure()
```

```python
df.plot(label='shanghai')
plt.legend(labels=['CD','BC','CD','D'],loc='best')
df.to_csv('foo.csv')
```
